```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10,10, num=100)

fx = []
for i in range(len(x)):
    fx.append(x[i]**2 - 2*x[i] + 5)
    
plt.plot(x, fx)
plt.grid()
plt.axvline()
plt.axvline()
plt.show()
```

```python
x = np.linspace(-10,10, num=100)

fx = []
for i in range(len(x)):
    fx.append(x[i]**2)
    
plt.plot(x, fx)
plt.grid()
plt.axvline()
plt.axvline()
plt.show()

```

```python
x = np.linspace(-5,5,100)
y = 1/(100 + x**6)
plt.plot(x, y)
plt.grid()
plt.axvline()
plt.show()
```

```python
range_min = -110
range_max = 110
poly = 15 
attr = 'Age'

plt.rcParams["figure.figsize"] = (8, 4)
x = np.linspace(range_min, range_max, 100)
y = (1/(poly*10 + x**2)) * poly * 10
plt.plot(x, y)
plt.grid()
plt.axvline()

plt.xlabel('Difference in ' + attr)
plt.ylabel('Similarity')
plt.title('Polynominal Similarity Function for ' + attr)

plt.show()
```
